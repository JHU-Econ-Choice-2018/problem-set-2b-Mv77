### What is the sustainable level of consumption?

We saw in [PerfectForesightCRRA](http://www.econ2.jhu.edu/people/ccarroll/public/lecturenotes/Consumption/PerfForesightCRRA) that to leave an agent's wealth untouched he must consume a fraction $$k = \frac{r}{R}$$ of his overall wealth $o_t = m_t - p_t + h_t$. As the problem is normalized by permanent income in this case we would have $p_t=1$ and $h_t = 1/(1-G/R)$. Therefore, the sustainable level of consumption will be $$\bar{c_t} = \frac{r}{R}(m_t - 1 + \frac{1}{1-G/R})$$

```python
# QUESTION: Now calculate and plot the "sustainable" level of consumption that leaves wealth untouched
# and plot it against the perfect foresight solution

human_wealth = 1/(1-(PermGroFac[0]/Rfree))
kappa = rfree/Rfree

cSustainable = kappa*(m_range-1+human_wealth) # For any given level of m, the level of c that would leave wealth unchanged
# Obviously, 0 is the wrong formula here -- you should fill in the right one

plt.figure(figsize=(12,9)) # set the figure size
plt.xlabel('Market resources m') # x axis label
plt.ylabel('Consumption c') # y axis label

plt.plot(m_range, cSustainable  , 'k', label='Sustainable c') # Add sustainable c line
plt.plot(m_range, cHARK, 'b', label='c Function')
plt.legend()

plt.show() # show the plot
```

### Patience factors

Absolute patience factor: $(R\beta)^{1/\rho}$

Return patience factor: $\frac{(R\beta)^{1/\rho}}{R}$ thus, for the RIC to hold, $\beta<R^{\rho-1}$

Growth patience factor: $\frac{(R\beta)^{1/\rho}}{G}$ thus, for the GIC to hold, $(R\beta)^{1/\rho}<G$



```python
# Compute the values of the impatience conditions under default parameter values

Pat_df  = (Rfree*DiscFac)**(1/CRRA) # Plug in the formula for the absolute patience factor
PatR_df = (Rfree*DiscFac)**(1/CRRA)/Rfree # Plug in the formula for the return patience factor
PatG_df = (Rfree*DiscFac)**(1/CRRA)/PermGroFac[0] # Plug in the formula for the growth patience factor

DiscFac_lim = Rfree**(CRRA - 1) # The limiting value such that the RIC exactly fails
PermGroFac_lim = (Rfree*DiscFac)**(1/CRRA) # The limiting value such that the GIC exactly fails

print('The absolute patience factor is %1.2f' % (Pat_df))
print('The return patience factor is %1.2f' % (PatR_df))
print('The growth patience factor is %1.2f' % (PatG_df))

print('The limiting value of Beta such that the RIC exactly fails is %.2f' % (DiscFac_lim))
print('The limiting value of G such that the GIC exactly fails is %.2f' % (PermGroFac_lim))
```

```python
# The code below is an example to show you how to plot a set of consumption functions
# for a sequence of values of the discount factor.  You should be
# to adapt this code to solve the rest of the sproblem posed above

howClose=0.001 # How close to come to the limit where the impatience condition fails
DiscFac_min = 0.8
DiscFac_max = DiscFac_lim-howClose # 
numPoints = 10
DiscFac_list = np.linspace(DiscFac_min, DiscFac_max, numPoints) # Create a list of beta values

plt.figure(figsize=((12,9))) # set the plot size

plt.plot(m_range, cSustainable  , 'k', label='Sustainable c') # Add sustainable c line
for i in range(len(DiscFac_list)):
    PFagent.DiscFac = DiscFac_list[i]
    PFagent.solve()
    cHARK = PFagent.solution[0].cFunc(m_range)
    plt.plot(m_range, cHARK, label='Consumption Function, $\\beta$= %1.2f' %(PFagent.DiscFac) )

PFagent.DiscFac = Params.DiscFac # return discount factor to default value
PFagent.solve() # It's nice to leave the PFagent back with its default solution
plt.xlabel('Market resources m') # x axis label
plt.ylabel('Consumption c')      # y axis label
plt.legend()                     # show legend
plt.show()                       # plot chart
```

The graphic shows that as the consumer's discount factor raises, his consumption flattens and moves towards zero. This is due to the fact that as $\beta$ rises, the consumer assigns more importance to his future consumption and thus saves more. The limiting behavior can be understood from the infinite horizon solution

\begin{equation}
c_t = \left( \frac{R - (R\beta)^{1/\rho}}{R} \right) o_t
\end{equation}

1. As $\beta\rightarrow 0$, then $c_t \rightarrow o_t$, showing that, as the consumer does not care about the future, he will consume all his wealth in the present.

2. As $\beta$ approaches the limit value that makes the RIC hold, we obtain $c_t \rightarrow \left(\frac{R - R}{R} \right) o_t = 0$. The consumer becomes so patient that he almost does not consume any fraction of his wealth initially. If the RIC is violated, the consumer is so patient, that he will always save all of his wealth, accumulating it for the end of times (which will never come).

## The consumption rule for varying G

We saw that for the GIC to hold, we need $(R\beta)^{1/\rho}<G$. However, for the present value of human wealth to be finite, permanent income must grow at a rate below the interest rate (this is the FHWC!). The maximum possible value for $G$ is therefore $R$. Therefore, we consider values of $G$ for which $$(R\beta)^{1/\rho}<G<R.$$

```python
# Now plot the consumption functions for alternate values of G as described above
# Note the tricky fact that PermGroFac is a list of values because it could 
# be representing some arbitrary sequence of growth rates
# QUESTION: What is the upper bound for possible values of G?  Why?

PermGro_min = PermGroFac_lim + howClose
PermGro_max = Rfree - howClose# Arbitrary number larger than the default value
PermGroArray = np.linspace(PermGro_min, PermGro_max, numPoints, endpoint=True)
PermGroList = PermGroArray.tolist() # Make growth factors a list

plt.figure(figsize=((12,9))) # set the plot size

plt.plot(m_range, cSustainable  , 'k', label='Sustainable c') # Add sustainable c line
for i in range(len(PermGroList)):
    PFagent.PermGroFac = [PermGroList[i]]
    PFagent.solve()
    cHARK = PFagent.solution[0].cFunc(m_range)
    plt.plot(m_range, cHARK, label='Consumption Function, $G$= %1.3f' %(PFagent.PermGroFac[0]) )

PFagent.PermGroFac = Params.PermGroFac # return discount factor to default value
PFagent.solve() # It's nice to leave the PFagent back with its default solution
plt.xlabel('Market resources m') # x axis label
plt.ylabel('Consumption c')      # y axis label
plt.legend()                     # show legend
plt.show()                       # plot chart
```

The effect of $G$ is to modify the consumer's human wealth, as we can see rearranging the consumption function:

\begin{align}
c_t &= \left( \frac{R - (R\beta)^{1/\rho}}{R} \right) o_t = \kappa o_t\\
&= \kappa \left(m_t-p_t+\frac{p_t}{1-G/R}\right) = \kappa m_t + \kappa \left(-p_t+\frac{p_t}{1-G/R}\right)
\end{align}

the expression makes clear that:
1. $c_t$ increases with $G$.
2. The increases happen as vertical shifts of the function $c_t(m_t)$, as only the intercept is changing.

The plot's behavior as $\rho$ changes can be understood from the consumption formula derived in class:

\begin{equation}
c_t = \left( \frac{R - (R\beta)^{1/\rho}}{R} \right) o_t
\end{equation}

and the fact that $R\beta = 0.99<1$.

As $\rho$ grows, $c_t \rightarrow \frac{R-1}{R}o_t = \frac{r}{R}o_t$, which is exactly the sustainable consumption. Intuitively, as the consumer becomes more and more inelastic, he will want to consume the same ammount every period, which is achieved by leaving his wealth constant.

Conversely, as $\rho$ goes to zero $c_t \rightarrow \frac{R-0}{R}o_t = o_t$, which tells us that the agent will consme all of his overall wealth in the first period. This is due to the fact that for this parametrization, the agent is impatient ($R\beta<1$) and so, as he becomes more willing to transer consumption through time, he will consume a larger portion of his overall wealth in the current period.
